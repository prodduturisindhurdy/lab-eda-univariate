```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

```python
df = pd.read_csv(r"C:\Users\prodd\Downloads\archive\amz_uk_price_prediction_dataset.csv")
df
```

```python
frequency_table=df["category"].value_counts()
frequency_table
```

```python
top_5_categories=frequency_table.sort_values(ascending=False).head(5)
top_5_categories
```

```python
sns.barplot(x=top_5_categories.index, y=top_5_categories.values,color='pink')
plt.xticks(rotation=45, ha='right')
plt.xlabel('category')
plt.ylabel('count')
plt.show()
```

```python
top_5_categories.plot.pie(autopct='%1.1f%%', startangle=40, colors=sns.color_palette("Set3"))
plt.show()
```

```python
mean_price=df["price"].mean()
median_price=df["price"].median()
mode_price=df["price"].mode()[0]
```

```python
mean_price,median_price,mode_price
```

```python
print(f"the average price of products listed is {mean_price}. In comparison with the {mode_price} . the average is significantly higher almost 10x. This tells us that there 9.99 is the most common price but still, considering all the products the average is higher despite that fact.")
```

```python
var_price=df["price"].var()
std_price=df["price"].std()
range_price=df["price"].max()-df["price"].min()
quartile_price=df["price"].quantile([0.25,0.5,0.75])
interquartile_price=quartile_price[0.75]-quartile_price[0.25]
```

```python
var_price,std_price,range_price,interquartile_price
```

```python
print("The prices are very varied. High variance, high std deviation, the range is wide, the mean is far away from the max value. It seems 50% of products are priced below 19.09, and the higher end of the quartiles appear to be more spreaded, more variance in the top direction.")
```

```python
sns.histplot(df['price'], kde=True, bins=30, color="salmon")
plt.show()
```

```python
print("After reviewing the histogram, I observed that most product prices are concentrated at the lower end, with a sharp drop-off as prices increase, indicating a right-skewed distribution and a significant spread driven by a small number of high-priced items.")
```
